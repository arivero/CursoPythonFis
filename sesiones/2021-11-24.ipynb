```python
from pyspark import SparkContext

sc = SparkContext("local[*]")
 
def mult(x):
    y = np.array([2,3,4])
    return x*y
 
x = np.arange(10000)
rdistData = sc.parallelize(x) #r is for resilient
 
results = rdistData.map(mult).collect()

print(results[-10:])
print(rdistData.map(mult).take(5))
sc.stop()
```

```python
sc = SparkContext("local[*]")
 
def long(x):
    return np.arange(x)
 
x = np.arange(10000)
rdistData = sc.parallelize(x) #r is for resilient
  
print(rdistData.flatMap(long).take(50))
sc.stop()
```

```python
#esto es intereante verlo en el htop
# 
sc.stop()
sc = SparkContext("local[*]")
 
def long(x):
    return np.arange(x)
 
x = np.arange(100000)
np.random.shuffle(x)
rdistData = sc.parallelize(x) #r is for resilient

#lo normal es hacer from operator import add
def add(x,y):
    return x+y
  
print(rdistData.flatMap(long).map(lambda x: x**2 ).reduce(add))
sc.stop()
```
