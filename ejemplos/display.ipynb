```python
#ojo, see https://ipycanvas.readthedocs.io/en/latest/installation.html#jupyterlab-extension
from ipycanvas import RoughCanvas, hold_canvas
```

```python
canvas = RoughCanvas(width=400, height=200) #esto es para que parezca hecho a mano 
```

```python
canvas
```

```python
canvas.fill_style = '#0000FF'
canvas.fill_rect(10, 10,120, 190)
```

```python
canvas.stroke_style = 'red'
canvas.stroke_line(200, 0, 0, 200)
```

```python
n = 50
x = np.linspace(0, 200, n)
y = np.random.randint(200, size=n)

points = np.stack((x, y), axis=1)

canvas.stroke_lines(points)
```

```python
canvas.begin_path()
canvas.move_to(75, 50)
canvas.line_to(100, 75)
canvas.line_to(100, 25)
canvas.fill()
```

```python
t=np.linspace(0, 20, 100)
x=t-10
y=t
z=t*t
#import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(x,y,z)
```

```python
###
### 
###
x,y,z = np.zeros((3,50,50))
for i in range(50):
    for j in range(50):
        x[i,j]=i/10
        y[i,j]=j/10
        z[i,j]=np.sqrt(i*j)
```

```python
np.shape(x)
```

```python
fig=plt.figure()
matplotlib.pyplot.contour(x,y,z)
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(x,y,z)
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z,cmap='viridis')
```
