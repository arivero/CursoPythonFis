```python
#ojo, see https://ipycanvas.readthedocs.io/en/latest/installation.html#jupyterlab-extension
from ipycanvas import RoughCanvas, hold_canvas
```

```python
canvas = RoughCanvas(width=400, height=200) #esto es para que parezca hecho a mano 
```

```python
canvas
```

```python
canvas.fill_style = '#0000FF'
canvas.fill_rect(10, 10,120, 190)
```

```python
canvas.stroke_style = 'red'
canvas.stroke_line(200, 0, 0, 200)
```

```python
n = 50
x = np.linspace(0, 200, n)
y = np.random.randint(200, size=n)

points = np.stack((x, y), axis=1)

canvas.stroke_lines(points)
```

```python
canvas
```

```python
canvas.begin_path()
canvas.move_to(75, 50)
canvas.line_to(100, 75)
canvas.line_to(100, 25)
canvas.fill()
```
